```python
import numpy as np
import matplotlib.pyplot as plt

from psi import *
from sigmoid import *

import warnings
warnings.filterwarnings('ignore')
```

### Sprecher's corrected $\psi$ function and its derivatives

```python
n = 2
g = 10
```

```python
k=1
k1_dk_span = get_Dk_span(0., 1, k, g)
k1_psi_span = get_psi_span(k1_dk_span, k, g)

k=2
k2_dk_span = get_Dk_span(0., 1, k, g)
k2_psi_span = get_psi_span(k2_dk_span, k, g)

k=3
k3_dk_span = get_Dk_span(0., 1, k, g)
k3_psi_span = get_psi_span(k3_dk_span, k, g)

k=4
k4_dk_span = get_Dk_span(0., 1, k, g)
k4_psi_span = get_psi_span(k4_dk_span, k, g)
```

#### Comparing $\psi(x)$ for different values of parameter $k$.

```python
fig, ax = plt.subplots(1,1,figsize=(8,8))
ax.plot(k1_dk_span, k1_psi_span, c='black', label='k=1')
ax.plot(k2_dk_span, k2_psi_span, c='red', label='k=2')
ax.plot(k3_dk_span, k3_psi_span, c='green', label='k=3')
ax.plot(k4_dk_span, k4_psi_span, c='orange', label='k=4')
ax.set(ylabel=r'$\psi(x)$', xlabel=r'$x$')
plt.legend()
plt.savefig('figs/psi_k.png');
```

```python
fig, ax = plt.subplots(1,4,figsize=(16,4))
ax[0].scatter(k1_dk_span, k1_psi_span, c='black', label='k=1', s=1)
ax[0].legend()
ax[1].scatter(k2_dk_span, k2_psi_span, c='red', label='k=2', s=1)
ax[1].legend()
ax[2].scatter(k3_dk_span, k3_psi_span, c='green', label='k=3', s=1)
ax[2].legend()
ax[3].scatter(k4_dk_span, k4_psi_span, c='orange', label='k=4', s=1)
ax[3].legend()
plt.savefig('figs/psi_k_scatter.png');
```

#### Studying derivatives of $\psi(x)$ for $k=4$.

```python
k=4

dk_span = get_Dk_span(0., 1, k, g)
psi_span = get_psi_span(dk_span, k, g)

psi = make_function(dk_span, psi_span)
psi_ = make_function(psi_span, dk_span)

d1psi = make_dfunction(psi, g**(-k), 1)
d2psi = make_dfunction(psi, g**(-k), 2)
```

```python
fig, axs = plt.subplots(1,3,figsize=(12,4))
axs[0].plot(dk_span, [psi(dk) for dk in dk_span], c='black', label=r"$\psi(x)$")
axs[0].legend(loc='upper left', prop={'size': 12})
axs[0].set(xlabel=r"$x$")
axs[1].plot(dk_span, [d1psi(dk) for dk in dk_span], c='black', label=r"$\psi'(x)$")
axs[1].legend(loc='upper left', prop={'size': 12})
axs[1].set(xlabel=r"$x$")
axs[2].plot(dk_span, [d2psi(dk) for dk in dk_span], c='black', label=r"$\psi''(x)$")
axs[2].legend(loc='upper left', prop={'size': 12})
axs[2].set(xlabel=r"$x$")

plt.tight_layout(0.5)
plt.savefig('figs/k4_psi_der.png');
```

### Lipschitz continuous $\psi$ function and its derivatives

#### Sprecher's $\psi$ vs. Lipschitz continuous

```python
k=2

k2_dk_span = get_Dk_span(0., 1., k, g)
k2_psi_lip_span = get_psi_lip_span(k2_dk_span, k, g)

k=3

k3_dk_span = get_Dk_span(0., 1., k, g)
k3_psi_lip_span = get_psi_lip_span(k3_dk_span, k, g)

k=4

k4_dk_span = get_Dk_span(0., 1., k, g)
k4_psi_lip_span = get_psi_lip_span(k4_dk_span, k, g)
```

```python
fig, axs = plt.subplots(1,3,figsize=(12,4))
axs[0].plot(k2_dk_span, k2_psi_lip_span, c='black', label='Lipschitz, k=2')
axs[0].plot(k2_dk_span, k2_psi_span, c='orange', label='Sprecher, k=2')
axs[0].set(xlabel=r"$x$",ylabel=r"$\psi(x), \tilde\psi(x)$")
axs[0].legend(prop={'size': 12})

axs[1].plot(k3_dk_span, k3_psi_lip_span, c='black', label='Lipschitz, k=3')
axs[1].plot(k3_dk_span, k3_psi_span, c='orange', label='Sprecher, k=3')
axs[1].set(xlabel=r"$x$")
axs[1].legend(prop={'size': 12})


axs[2].plot(k4_dk_span, k4_psi_lip_span, c='black', label='Lipschitz, k=4')
axs[2].plot(k4_dk_span, k4_psi_span, c='orange', label='Sprecher, k=4')
axs[2].set(xlabel=r"$x$")
axs[2].legend(prop={'size': 12})

plt.tight_layout(0.5)

plt.savefig('figs/sprech_vs_lip.png');
```

#### Derivatives of Lipschitz continuous $\psi$ function

```python
k=4

dk_span = get_Dk_span(0., 1., k, g)
psi_lip_span = get_psi_lip_span(dk_span, k, g)

psi_lip = make_function(dk_span, psi_lip_span)
psi_lip_ = make_function(psi_lip_span, dk_span)

d1psi_lip = make_dfunction(psi_lip, g**(-k), 1)
d2psi_lip = make_dfunction(psi_lip, g**(-k), 2)
d3psi_lip = make_dfunction(psi_lip, g**(-k), 3)
```

```python
fig, axs = plt.subplots(1,3,figsize=(12,4))
axs[0].plot(dk_span, [psi_lip(dk) for dk in dk_span], c='black', label=r"$\tilde\psi(x)$")
axs[0].legend(loc='upper left', prop={'size': 12})
axs[0].set(xlabel=r"$x$")
axs[1].plot(dk_span, [d1psi_lip(dk) for dk in dk_span], c='black', label=r"$\tilde\psi'(x)$")
axs[1].legend(loc='upper left', prop={'size': 12})
axs[1].set(xlabel=r"$x$")
axs[2].plot(dk_span, [d2psi_lip(dk) for dk in dk_span], c='black', label=r"$\tilde\psi''(x)$")
axs[2].legend(loc='upper left', prop={'size': 12})
axs[2].set(xlabel=r"$x$")

plt.tight_layout(0.5)
plt.savefig('figs/k4_psi_lip_der.png');
```

### Sigmoid approximation of $\psi$ function

#### $\sigma(x)$ vs $\psi(x)$

```python
psi_sigmoid = PSI_sigmoid()
sigma = psi_sigmoid.psi
sigma_ = psi_sigmoid.psi_
```

```python
fig,ax = plt.subplots(1,1,figsize=(6,6))

x_span = get_Dk_span(-1, 1, 4, 10)
ax.plot(x_span, [get_psi(x) for x in x_span], c='orange', label=r"$\psi(x)$")
ax.plot(x_span, [sigma(x) for x in x_span], c='black', label=r"$\sigma(x)$")
ax.set(xlabel='x')
ax.legend()
plt.savefig('figs/sigma_vs_psi.png');
```

#### Derivatives of $\sigma(x)$

```python
d1sigma = psi_sigmoid.d1psi
d2sigma_ = psi_sigmoid.d2psi
```

```python
fig, axs = plt.subplots(1,3,figsize=(12,4))
axs[0].plot(x_span, [sigma(x) for x in x_span], c='black', label=r"$\sigma(x)$")
axs[0].legend(loc='upper left', prop={'size': 12})
axs[0].set(xlabel=r"$x$")
axs[1].plot(x_span, [d1sigma(x) for x in x_span], c='black', label=r"$\sigma'(x)$")
axs[1].legend(loc='upper left', prop={'size': 12})
axs[1].set(xlabel=r"$x$")
axs[2].plot(x_span, [d2sigma_(x) for x in x_span], c='black', label=r"$\sigma''(x)$")
axs[2].legend(loc='upper left', prop={'size': 12})
axs[2].set(xlabel=r"$x$")

plt.tight_layout(0.5)
plt.savefig('figs/k4_sigma_der.png');
```
